## OOP WRAPPER✔

```python
# Base Class
class Employee:

    # Constructor
    def __init__(self, emp_id=0, name="Unknown", age=0, salary=0):
        self.__emp_id = emp_id      # Private variable
        self.name = name
        self.age = age
        self.__salary = salary

    # Getter for Employee ID
    def get_emp_id(self):
        return self.__emp_id

    # Setter for Employee ID
    def set_emp_id(self, emp_id):
        self.__emp_id = emp_id

    # Getter for Salary
    def get_salary(self):
        return self.__salary

    # Setter for Salary
    def set_salary(self, salary):
        self.__salary = salary

    # Display Method
    def display(self):
        print("\nEmployee Details")
        print("Employee ID :", self.__emp_id)
        print("Name        :", self.name)
        print("Age         :", self.age)
        print("Salary      :", self.__salary)

    # Destructor
    def __del__(self):
        print("Employee object deleted")


# Derived Class - Manager
class Manager(Employee):

    def __init__(self, emp_id, name, age, salary, department):
        super().__init__(emp_id, name, age, salary)
        self.department = department

    # Method Overriding
    def display(self):
        super().display()
        print("Department  :", self.department)


# Derived Class - Developer
class Developer(Employee):

    def __init__(self, emp_id, name, age, salary, language):
        super().__init__(emp_id, name, age, salary)
        self.language = language

    # Method Overriding
    def display(self):
        super().display()
        print("Language    :", self.language)


# Main Program
employees = []

while True:

    print("\n===== Employee Management System =====")
    print("1. Add Manager")
    print("2. Add Developer")
    print("3. Display Employees")
    print("4. Check Subclass")
    print("5. Exit")

    choice = input("Enter your choice : ")

    # Add Manager
    if choice == "1":

        emp_id = int(input("Enter Employee ID : "))
        name = input("Enter Name : ")
        age = int(input("Enter Age : "))
        salary = int(input("Enter Salary : "))
        department = input("Enter Department : ")

        manager = Manager(emp_id, name, age, salary, department)
        employees.append(manager)

        print("Manager Added Successfully!")

    # Add Developer
    elif choice == "2":

        emp_id = int(input("Enter Employee ID : "))
        name = input("Enter Name : ")
        age = int(input("Enter Age : "))
        salary = int(input("Enter Salary : "))
        language = input("Enter Programming Language : ")

        developer = Developer(emp_id, name, age, salary, language)
        employees.append(developer)

        print("Developer Added Successfully!")

    # Display Employees
    elif choice == "3":

        if len(employees) == 0:
            print("No Employee Records Found")

        else:
            for emp in employees:
                emp.display()

    # Check issubclass()
    elif choice == "4":

        print("\nManager is subclass of Employee :",
              issubclass(Manager, Employee))

        print("Developer is subclass of Employee :",
              issubclass(Developer, Employee))

    # Exit
    elif choice == "5":

        print("Program Closed")
        break

    else:
        print("Invalid Choice")
```
